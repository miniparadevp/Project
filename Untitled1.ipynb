```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
file_path = 'Training.csv'
data = pd.read_csv(file_path)

# Encode target labels
data['prognosis'] = data['prognosis'].astype('category').cat.codes

# Extract features and labels
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dimensionality Reduction (PCA)
pca = PCA(n_components=min(X.shape[1], 20))  # Keep max 20 important features
X_scaled = pca.fit_transform(X_scaled)

# Custom SVM Classifier
class SVCFromScratch:
    def __init__(self, learning_rate=0.01, lambda_param=0.01, n_iters=2000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))
        
        np.random.seed(42)  # 🔹 Ensure different initial weights per run
        self.weights = np.random.randn(n_classes, n_features) * 0.01  # Small random weights
        self.bias = np.random.randn(n_classes) * 0.01
        
        for c in range(n_classes):
            y_binary = np.where(y == c, 1, -1)  # Convert to binary for class c
            
            for _ in range(self.n_iters):
                margin = y_binary * (np.dot(X, self.weights[c]) - self.bias[c])
                misclassified = margin < 1
                
                self.weights[c] -= self.lr * (2 * self.lambda_param * self.weights[c] - np.dot(X.T, y_binary * misclassified))
                self.bias[c] -= self.lr * np.sum(y_binary * misclassified)

    def predict(self, X):
        scores = np.dot(X, self.weights.T) - self.bias
        return np.argmax(scores, axis=1)
    
    def get_params(self, deep=True):
        return {"learning_rate": self.lr, "lambda_param": self.lambda_param, "n_iters": self.n_iters}
    
    def set_params(self, **params):
        for key, value in params.items():
            setattr(self, key, value)
        return self


# Train the SVM model
model = SVCFromScratch()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = np.mean(y_pred == y_test) * 100
print(f"Improved Model Test Accuracy: {accuracy:.2f}%")
```

```python

# Cross-validation function with visualization
def cross_validate_with_visualization(model, X, y, k=5):
    skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)  # 🔹 Stratified ensures balanced class splits
    accuracies = []

    for fold, (train_index, test_index) in enumerate(skf.split(X, y), 1):
        X_train_fold, X_test_fold = X[train_index], X[test_index]
        y_train_fold, y_test_fold = y[train_index], y[test_index]

        model.fit(X_train_fold, y_train_fold)  
        y_pred_fold = model.predict(X_test_fold)  
        
        accuracy = np.mean(y_pred_fold == y_test_fold) * 100  
        accuracies.append(accuracy)
        print(f"Fold {fold} Accuracy: {accuracy:.2f}%")

    mean_accuracy = np.mean(accuracies)
    std_deviation = np.std(accuracies)

    print("\nFinal Cross-Validation Results:")
    print(f"Mean Accuracy: {mean_accuracy:.2f}%")
    print(f"Standard Deviation: {std_deviation:.2f}%")

    # Plotting results
    plt.figure(figsize=(14, 5))

    # 1️⃣ Box Plot & Violin Plot of Accuracy Distribution
    plt.subplot(1, 3, 1)
    sns.boxplot(y=accuracies, color="blue")
    sns.violinplot(y=accuracies, color="lightblue", alpha=0.5)
    plt.ylabel("Accuracy (%)")
    plt.title("Accuracy Distribution (Cross Validation)")

    # 2️⃣ Line Plot of Accuracy Across Folds
    plt.subplot(1, 3, 2)
    plt.plot(range(1, k + 1), accuracies, marker="o", linestyle="-", color="green", label="Fold Accuracy")
    plt.axhline(mean_accuracy, color="red", linestyle="dashed", label="Mean Accuracy")
    plt.xlabel("Fold Number")
    plt.ylabel("Accuracy (%)")
    plt.title("Cross Validation Accuracy Trend")
    plt.legend()

    # 3️⃣ Bar Plot of Mean Accuracy with Error Bars
    plt.subplot(1, 3, 3)
    plt.bar(["Mean Accuracy"], [mean_accuracy], yerr=[std_deviation], capsize=10, color="orange", alpha=0.7)
    plt.ylabel("Accuracy (%)")
    plt.title("Mean Accuracy with Standard Deviation")

    plt.tight_layout()
    plt.show()

    return mean_accuracy, std_deviation

# Run cross-validation with visualization
cross_validate_with_visualization(SVCFromScratch(learning_rate=0.05, lambda_param=0.001, n_iters=2000), X_scaled, y, k=5)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dataset
file_path = 'Training.csv'
data = pd.read_csv(file_path)

# Encode target labels
data['prognosis'] = data['prognosis'].astype('category').cat.codes

# Extract features and labels
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dimensionality Reduction (PCA)
pca = PCA(n_components=min(X.shape[1], 20))  # Keep max 20 features
X_scaled = pca.fit_transform(X_scaled)

# Define models for comparison
models = {
    "SVM (Custom)": SVCFromScratch(learning_rate=0.05, lambda_param=0.001, n_iters=2000),
    "SVM (Sklearn)": SVC(kernel="linear"),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "Logistic Regression": LogisticRegression()
}

# Perform cross-validation & collect metrics
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {}

for name, model in models.items():
    accuracies = []
    precisions = []
    recalls = []
    f1_scores = []

    for train_idx, test_idx in cv.split(X_scaled, y):
        X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        accuracies.append(accuracy_score(y_test, y_pred) * 100)
        precisions.append(precision_score(y_test, y_pred, average='weighted') * 100)
        recalls.append(recall_score(y_test, y_pred, average='weighted') * 100)
        f1_scores.append(f1_score(y_test, y_pred, average='weighted') * 100)

    results[name] = {
        "Accuracy": np.mean(accuracies),
        "Precision": np.mean(precisions),
        "Recall": np.mean(recalls),
        "F1-Score": np.mean(f1_scores)
    }

# Convert results to DataFrame
results_df = pd.DataFrame(results).T

# Print results
print("\nModel Performance Comparison:")
print(results_df)

# Visualization of Model Comparison
plt.figure(figsize=(12, 6))
results_df.plot(kind="bar", figsize=(12, 6), colormap="viridis", rot=45)
plt.title("Model Performance Comparison (Cross-Validation Metrics)")
plt.ylabel("Score (%)")
plt.legend(loc="best")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
